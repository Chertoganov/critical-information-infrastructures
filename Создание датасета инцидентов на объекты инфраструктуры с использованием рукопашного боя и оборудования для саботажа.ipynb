```python
# выбираем объекты инфраструктуры
gtd1993_0919dist[gtd1993_0919dist["attacktype1_txt"]=="Facility/Infrastructure Attack"]
```

```python
# создаем датафрейм объектов инфраструктуры
Facility = gtd1993_0919dist[gtd1993_0919dist["attacktype1_txt"]=="Facility/Infrastructure Attack"]
```

```python
# смотрим тип оружия
Facility["weaptype1_txt"].to_list()
```

```python
# группируем инциденты на объекты инфраструктуры по оружию
Facility.groupby('weaptype1_txt').count()
```

```python
# создрание датафрейма из инцидентов на объекты инфраструктуры, тип оружия - оборудование для саботажа
Facility_Sabotage_Equipment = Facility[Facility["weaptype1_txt"]=="Sabotage Equipment"]
Facility_Sabotage_Equipment
```

```python
# создрание датафрейма из инцидентов на объекты инфраструктуры, тип оружия - рукопашный бой
Facility_Melee = Facility[Facility["weaptype1_txt"]=="Melee"]
Facility_Melee
```
