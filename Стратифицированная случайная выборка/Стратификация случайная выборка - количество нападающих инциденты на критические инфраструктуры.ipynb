# Пример как работает стратифицированная случайная выборка

```python
import pandas as pd

#create DataFrame
df = pd.DataFrame({'team': ['A', 'A', 'B', 'B', 'B', 'B', 'B', 'B'],
                   'position': ['G', 'G', 'F', 'G', 'F', 'F', 'C', 'C'],
                   'assists': [5, 7, 7, 8, 5, 7, 6, 9],
                   'rebounds': [11, 8, 10, 6, 6, 9, 6, 10]})

#view DataFrame
df
```

```python
import numpy as np

#define total sample size desired
N = 4

#perform stratified random sampling
df.groupby('team', group_keys=False).apply(lambda x: x.sample(int(np.rint(N*len(x)/len(df))))).sample(frac=1).reset_index(drop=True)
```

```python
# загружаем чистый набор данных инцдентов на критические инфраструктуры с использованием рукопашного боя (32 инцидента)
df1 = pd.read_csv('Facility_Melee_clean_32_dataset.csv')
df1.head()
```

```python
# загружаем чистый набор данных инцдентов на критические инфраструктуры с использованием оборудования для саботажа (20 инцидентов)
df2 = pd.read_csv('Facility_Sabotage_Equipment_clean_20_dataset.csv')
df2.head()
```

```python
# создаем класс по которому будет происходить разделение
df2["fight"] = "саботаж"
```

```python
# смотрим класс по которому будет происходить разделение
df2["fight"]
```

```python
# размер чистыого набора данных инцидентов на критические инфраструктуры с использованием оборудования для саботажа (20 инцидентов)
df2.shape
```

```python
import numpy as np

# определяем размер выборки для первой группы (рукопашный бой)
N = 18

# делаем стратифицированную случайную выборку
a = df1.groupby('fight', group_keys=False).apply(lambda x: x.sample(int(np.rint(N*len(x)/len(df1))))).sample(frac=1).reset_index(drop=True)
a
```

```python
# смотрим стратифицированную случайную выборку
a.nperps
```

```python
import numpy as np

# определяем размер выборки для второй группы (саботаж)
N = 18

# делаем стратифицированную случайную выборку
b = df2.groupby('fight', group_keys=False).apply(lambda x: x.sample(int(np.rint(N*len(x)/len(df2))))).sample(frac=1).reset_index(drop=True)
b
```

```python
# смотрим стратифицированную случайную выборку
b.nperps
```
